```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# RGB Image

```python
img = cv2.imread("img/PROFILE.jpg")
cv2.namedWindow("output",cv2.WINDOW_NORMAL)
cv2.resizeWindow("output",600,600)
cv2.imshow("output",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
print("type(img):", type(img))
print("shape of RGB:", img.shape)
print("dtype of RGB:", img.dtype)
print("unique values in RGB:", np.unique(img))
print("unique values sum in RGB:", len(np.unique(img)))
```

# Black and White(0,1)(0,1)

```python
img_binary = cv2.imread("img/PROFILE.jpg",cv2.IMREAD_GRAYSCALE)
_, binary_img = cv2.threshold(img_binary,127,255,cv2.THRESH_BINARY)
binary_img_01 = binary_img / 255
cv2.imshow("output",binary_img_01*255)
print("binary shape:", binary_img_01.shape)
print("binary dtype:", binary_img.dtype)
print("unique values in binary:", np.unique(binary_img_01))
print("unique values in binary:", len(np.unique(binary_img)))

print("size (pixels):", binary_img.size)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Gray image

```python
cv2.namedWindow("output",cv2.WINDOW_NORMAL)
cv2.resizeWindow("output",600,600)
img_gray = cv2.imread("img/PROFILE.jpg",cv2.IMREAD_GRAYSCALE)
cv2.imshow("output",img_gray)
print("gray shape:", img_gray.shape)
print("binary shape:", img_gray.shape)
print("binary dtype:", img_gray.dtype)
print("unique values in binary:", np.unique(img_gray))
print("size (pixels):", img_gray.size)

plt.hist(img_gray,cmap='gray', bins=256)
plt.show()

cv2.waitKey(0)
cv2.destroyAllWindows()
```
